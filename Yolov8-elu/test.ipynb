```python
import os
import cv2
from ultralytics import YOLO

def process_images(input_dir, output_dir, model_path, threshold=0.25):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    model = YOLO(model_path)
    jpg_files = [f for f in os.listdir(input_dir) if f.lower().endswith(".jpg")]

    for filename in jpg_files:
        image_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)

        image = cv2.imread(image_path)
        if image is None:
            print(f"âŒ Error: Could not read {image_path}")
            continue

        results = model(image)[0]

        if not results.boxes:
            print(f"âš ï¸ No objects detected in {filename}")
        else:
            for box in results.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = float(box.conf[0])
                cls = int(box.cls[0])

                if conf > threshold:
                    label = f"{model.names[cls]}: {conf:.2f}"

                    # Draw bounding box
                    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

                    # Get text size
                    font = cv2.FONT_HERSHEY_SIMPLEX
                    font_scale = 0.6
                    thickness = 1
                    (text_w, text_h), _ = cv2.getTextSize(label, font, font_scale, thickness)

                    # Determine label position (top or bottom of box)
                    label_y = y1 - 10 if y1 - 10 > text_h else y2 + text_h + 10

                    # Draw label background
                    cv2.rectangle(image, (x1, label_y - text_h), (x1 + text_w, label_y), (0, 255, 0), -1)
                    # Put label text
                    cv2.putText(image, label, (x1, label_y - 2), font, font_scale, (0, 0, 0), thickness)

                    # Console output
                    print(f"ğŸ”¹ {filename} - Class {cls} ({model.names[cls]}), Conf: {conf:.2f}, Box: ({x1}, {y1}, {x2}, {y2})")

        cv2.imwrite(output_path, image)
        print(f"âœ… Saved: {output_path}")
```

```python
model_path = '/content/drive/MyDrive/Yolov8-elu/best.pt'
input_directory = '/content/drive/MyDrive/Yolov8-elu/preprocess'
output_directory = '/content/drive/MyDrive/Yolov8-elu/processed_images/test8'

process_images(input_directory, output_directory, model_path)
```

```python
import glob
import matplotlib.pyplot as plt
from PIL import Image

# Get the list of image paths
image_paths = glob.glob('/content/drive/MyDrive/Yolov8-elu/processed_images/test7/*.jpg')[0:189]

# Set the number of images per row
num_cols = 3  # 3 images per row
num_rows = (len(image_paths) + num_cols - 1) // num_cols  # Calculate required rows

# Set up the plot with a larger figure size for better visibility
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
axes = axes.flatten()  # Flatten for easy iteration

# Display images in the grid
for idx, image_path in enumerate(image_paths):
    img = Image.open(image_path)
    axes[idx].imshow(img)
    axes[idx].axis('off')  # Hide axis

# Remove any unused subplots
for i in range(len(image_paths), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
